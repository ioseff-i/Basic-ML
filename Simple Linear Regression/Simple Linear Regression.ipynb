# Simple Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Import the dataset

```python
data_set = pd.read_csv('Salary_Data.csv')
X = data_set.iloc[:,:-1]
y = data_set.iloc[:,-1]
```

## Train Test split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

## Train the SLR on training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

## Obtained Parameters 

```python
print(regressor.coef_)
print(regressor.intercept_)
```

## Predicting the test set results

```python
y_pred_test = regressor.predict(X_test)
```

## Visualization of training set results

```python
plt.scatter(X_train,y_train,color='red')
plt.plot(X_train,regressor.predict(X_train),color='green')
plt.title('Experience vs Salary (training)')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```

## Visualization of test set results

```python
plt.scatter(X_test,y_test,color='blue')
plt.plot(X_test,y_pred_test,color='green')
plt.title('Experience vs Salary (testing)')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```
